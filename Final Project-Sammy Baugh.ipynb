### Section I: Finding and Cleaning the Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#https://www.pro-football-reference.com/players/B/BaugSa00.htm
stats = pd.read_csv('SlinginSammy.csv',index_col = 'Year')
stats
```

```python
stats = stats.drop(['GS','Y/A','AY/A','Y/C','Y/G'],axis=1)
stats = stats.drop(['Care'])
stats
```

### Section II: Elementary Data Analysis

```python
stats.G.mean()
```

One of the things that makes it difficult to compare players from the past and the present is that the NFL has changed a lot since its creation. For example, the number of games in a season was only 12 during Sammy Baugh's career, compared to 16 games now. Additionally, during World War II the seasons were further shortened. Samy Baugh's stats may seem insignificant compared to modern day QB's, but when you realize that he played an average of just 10.3 games a season, his stats seem much more impressive.

```python
stats.Yds.std()
```

Aside from pioneering the forward pass in the NFL, Sammy Baugh did a great many other things as a player for the Washington Redskins. In addition to playing Quarterback, he also played Tailback and Punter and Cornerback. The large standard deviation in his passing yards reflects the fact that he was not solely a Quarterback, but during different seasons he would perform other roles.

```python
stats.Lng.max()
```

During the early days of football, the game was predominantly a running game. Very few teams passed the ball, and those that did rarely threw it more than 30 yards. This is why Sammy Baugh's longest pass of 86 yards is significant. this is what earned him the nickname "Slingin' Sammy Baugh" and also what helped to usher in a new era of football where the forward pass was a much bigger part of the game.

### Section III: Special Pandas

```python
stats.groupby('G').max()
```

Grouping by the number of games played gives interesting results. The maximum number of yards and touchdowns in a given season does not seem to be affected by the number of games played. This is likely the result of the fact that Sammy Baugh did not always play at Quarterback, and so the Quarterback stats such as Yds and TDs depends more on what position he played in a season rather than how many games he played.

### Section IV: Matplotlib

```python
plt.plot(stats.Att.values,stats['TD'].values,'o')
plt.title('Sammy Baugh: TDs vs Attempts')
plt.xlabel('Pass Attempts')
plt.ylabel('Touchdowns')
```

This plot shows that more pass attempts results in more Touchdowns. Therefore, in order to win, Sammy should pass the ball more. Note: This is why Sammy's use of the forward pass was so revolutionary. He demonstrated that teams could win more games by throwing the ball, rather than running it, and this was a big turning point in converting the game of football into the pass heavy game it is today.

### Section VI: Sklearn

```python
import sklearn
```

```python
from sklearn.cluster import KMeans
```

```python
set1 = stats.Att
#set2 = stats.G
#X = np.concatenate([set1,set2])
X = np.concatenate([set1])
X = X[:,None]
scores = []
for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    score = -kmeans.score(X)
    scores.append(score)
plt.plot(np.arange(1,10),scores)
```

```python
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
preds = kmeans.predict(X)
preds
```

```python
pred0 = X[preds==0]
pred1 = X[preds==1]
pred2 = X[preds==2]
plt.scatter(pred0,pred0)
plt.scatter(pred1,pred1)
plt.scatter(pred2,pred2)
```

Based on the elbow rule, there appears to be 3 clusters in the data.
The clusters generally correspond to the number of games played. This can be valuable to the coach because it shows that if Baugh plays in more games he will throw the ball more, and as we have seen, more throws means more winning.
